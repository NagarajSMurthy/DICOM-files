```python
import pydicom
```

```python
file = "bmode.dcm"
ds = pydicom.dcmread(file)
```

```python
# Accessing individual data elements 
data_element = ds.data_element("PatientName")
print(data_element.VR)
print(data_element.tag)
```

```python
# To get the entire data elements
for data_elements in ds:
    print(data_elements.tag,"\t",data_elements.VR,"\t",data_elements.value)
```

```python
# To print the tag values
for data_elements in ds:
    print(data_elements.tag,"\n")
```

```python
# Writing to a text file
f = open("bmode.txt",'w')
f.write("Tags"+"\t\t"+"VR"+"\n")
for data_elements in ds:
    f.write(str(data_elements.tag)+"\t"+str(data_elements.VR)+"\n")
    
f.close()
```
